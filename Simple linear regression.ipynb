Simple linear regression : one output row vs one input row
hypothesis : y_hat = mx + b

```python
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('placement_data_SLR.csv')
dataset
```

```python
plt.scatter(dataset['cgpa'],dataset['package'],color='red') # setting the x,y axes for the dataSet X--> CGPA , Y--> Package
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
plt.show()
```

```python
x = dataset.iloc[:,0:1] # taking CGPA as x axis
y = dataset.iloc[:,-1] # taking salary package as y axis
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
SLR = LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
SLR.fit(x_train,y_train)
y_pred = SLR.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
r2_score
```

```python
plt.scatter(dataset['cgpa'],dataset['package'],color='red') # setting the x,y axes for the dataSet X--> CGPA , Y--> Package
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
m = SLR.coef_ # gives the slope of 'best fit line'
c = SLR.intercept_ # gives the intercept of 'best fit line'
y = m*x_train+c
plt.plot(x_train,y,color='green')
plt.show()
```
